# Thuật toán NSGA2

```
from pymoo.algorithms.nsga2 import  NSGA2
from pymoo.factory import get_problem, get_termination
from pymoo.util.termination.default import SingleObjectiveDefaultTermination
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
def Minimize(problem, algorithm,termination=None, seed = 17521056, verbose = False):
    res = minimize(problem,
               algorithm,
               termination = termination,
               seed=seed,
               verbose=verbose)
    return res

    
```

## unconstrained

### zdt1

```
# TH n_gen = 100
from pymoo.factory import get_performance_indicator

problem = get_problem("zdt1")
algorithm = NSGA2(pop_size=100)
termination = get_termination('n_gen',100)
res = Minimize(problem, algorithm, termination#,verbose=True
               )
pf = problem.pareto_front()
Scatter(legend=True).add(pf, label="Pareto-front").add(res.F, label="Result").show()

# GA dictance
gd = get_performance_indicator("gd", pf)
print("GD: ", gd.calc(res.F))

#  IGD 
igd = get_performance_indicator("igd", pf)
print("IGD: ", igd.calc(res.F))

# HV
hv = get_performance_indicator("hv", ref_point=np.array([1.0, 1.0]))
print("hv", hv.calc(res.F))
```
